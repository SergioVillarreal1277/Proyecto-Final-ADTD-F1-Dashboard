En este notebook cargamos los 12 archivos CSV de nuestra base de datos de Fórmula 1 
desde el Volume donde los subimos, y los registramos como tablas en Databricks 
para que los demás notebooks puedan usarlos.

```python
# Cargo los 12 CSVs desde el volume donde los subimos
# y los registro como tablas para poder consultarlos con SQL despues

# esta es la ruta base donde estan todos nuestros archivos en Databricks
ruta_base = "/Volumes/workspace/default/f1_datos"

# verifico que el volume existe y tiene archivos
try:
    archivos_en_volume = dbutils.fs.ls(ruta_base)
    if len(archivos_en_volume) == 0:
        raise FileNotFoundError(f"El volume '{ruta_base}' existe pero está vacío. Por favor sube los archivos CSV primero.")
    print(f"✓ Volume encontrado con {len(archivos_en_volume)} archivos\n")
except Exception as e:
    if "PATH_NOT_FOUND" in str(e):
        raise FileNotFoundError(f"El volume '{ruta_base}' no existe. Por favor créalo y sube los archivos CSV.")
    else:
        raise

# lista con los nombres de los archivos y como queremos llamar cada tabla
tablas = [
    ("Datos_Temporada.csv",    "temporada"),
    ("Datos_Circuito.csv",     "circuito"),
    ("Datos_Escuderia.csv",    "escuderia"),
    ("Datos_Patrocinador.csv", "patrocinador"),
    ("Datos_Piloto.csv",       "piloto"),
    ("Datos_Ingeniero.csv",    "ingeniero"),
    ("Datos_Carrera.csv",      "carrera"),
    ("Datos_Auto.csv",         "auto"),
    ("Datos_Vuelta.csv",       "vuelta"),
    ("Datos_Resultado.csv",    "resultado"),
    ("Datos_Tiempo_Vuelta.csv","tiempo_vuelta"),
    ("Datos_Financiamiento.csv","financiamiento"),
]

# recorro cada archivo, lo leo y lo registro como tabla
for archivo, nombre_tabla in tablas:
    ruta_completa = f"{ruta_base}/{archivo}"
    
    # leo el CSV con spark, que es el motor de datos de Databricks
    # header=True significa que la primera fila tiene los nombres de las columnas
    # inferSchema=True hace que Databricks detecte automaticamente el tipo de cada columna
    df = spark.read.csv(ruta_completa, header=True, inferSchema=True)
    
    # registro la tabla para poder consultarla con SQL en cualquier notebook
    df.write.mode("overwrite").saveAsTable(nombre_tabla)
    
    print(f"Tabla '{nombre_tabla}' cargada con {df.count()} filas")

print("\nTodas las tablas fueron cargadas")
```
